```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array((1,5))
y = np.array((1,5))

plt.plot(x,y)
plt.show()
```

```python
x = np.array(["futa","Bake","voli","hockey"])
y = np.array([30,15,20,25])

plt.bar(x,y,color="green")
plt.show()
```

```python
x =np.array([80,40,100,90])
my_label =["bio","math","english","swahili"]
my_colors=["brown","grey","green","black"]
my_explode=[0.3,0.1,0.2,0.1]

plt.pie(x,labels=my_label,colors=my_colors,explode=my_explode,shadow=True)
plt.legend(title="genius")
plt.show()
```
