```python
import numpy as np

a =np.array(6)
b =np.array([2,6,8,5])
c =np.array([[2,4,6],[6,6,7]])
d =np.array([[[4,6,7],[4,6,8]],[[5,7,8],[6,7,8]]])

print(b[1])
print("sum:",b[2]+b[2]+b[1])

print("zero dimentional:",a)
print("1-D:",b)
print("2-D:",c)
print("3-D:",d)

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)

print(c[1,2])
print(d[0,0,1])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2,3,4])
y = np.array([0,1,2,3,4])

plt.plot(x,y,marker = "*")
plt.title("my area")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.grid()
plt.show()
```

```python
x_axis =np.array([0,3,5,7])
y_axis =np.array([0,5,8,16])

plt.scatter(x_axis,y_axis)
plt.show()
```

```python
x = np.array(["Aples","Bananas","mango","grapes"])
y = np.array([4,15,20,2])

plt.bar(x,y,color="black")
plt.show()
```

```python
x = np.array(["Aples","Bananas","mango","grapes"])
y = np.array([4,15,20,2])

plt.barh(x,y,color="black",height=0.5)
plt.show()
```

```python
x =np.array([70,50,10,90])
my_label =["BMW","lambp","jeep","hammer"]
my_colors=["brown","grey","green","black"]
my_explode=[0.3,0.4,0.5,0.3]

plt.pie(x,labels=my_label,colors=my_colors,explode=my_explode,shadow=True)
plt.legend(title="genius")
plt.show()
```
